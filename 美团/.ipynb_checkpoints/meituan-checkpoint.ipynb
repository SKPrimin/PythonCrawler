## 数据分析
#### 相关模块导入 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['KaiTi']  # 指定默认字体
```

#### 文件数据读入
使用pandas读取信息并按列分开

```python
df = pd.read_csv('奶茶data.csv')
df.columns = ['店铺名', '店铺所在位置', '人均消费', '评分', '美食名称', '店铺图片链接', 
              '纬度', '经度', '最低价格', '店铺ID', '店铺详情页']
# print(df)
```

#### 价格与评分折线图
通过画出人均价格与评分之间的关系，我们可以清楚的观察到是否越贵的东西越好,同时也可以引入最低价格

```python
def list_reset_zero(alist):
    '''去除列表中0 的函数，去除数据中心未设置的部分，为了尽量不影响图像，将赋予其前一个值'''
    before = 0.0
    for i in range(len(alist)):
        if alist[i] == 0.0:
            alist[i] = before
        before =  alist[i]
```

```python
# ascending=False设置为降序排列
df.sort_values(by=['评分'], ascending=False, inplace=True)
# make data
x = df['店铺名']
y1 = df['人均消费']
y2 = df['评分']
y3 = df['最低价格']

list_reset_zero(y1)
list_reset_zero(y3)

# plot
fig, ax = plt.subplots(figsize=(18,9))

ax.plot(x, y1,color = 'aquamarine',linestyle='-', linewidth=2.0)
ax.plot(x, y2,color = 'dodgerblue',linestyle=':', linewidth=2.0)
ax.plot(x, y3,color = 'tomato',linestyle='-.', linewidth=2.0)


ax.set( xticks=x,
       ylim=(0, 20), yticks=np.arange(1, 20))
plt.xticks(rotation=30)
plt.show()
```

```python
#### 出现品牌统计
import re
brand = []
for item in df['店铺名']:
    i = re.findall("(.*?)\（", item)
    brand.append(i.)

print(brand)
```
